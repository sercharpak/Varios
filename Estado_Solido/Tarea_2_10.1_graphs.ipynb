```python
%pylab inline
import numpy as np
import matplotlib as plt
```

```python
#Omega
def omega_plus(kappa, m1, m2, k, a):
    raiz= kappa*((m1+m2)*(m1+m2)-4.0*(m1*m2)*((np.sin(k*a/2.0))**2.0))**(0.5)
    denum=m1*m2
    num=kappa*(m1+m2)+raiz
    fraccion=num/denum
    return (fraccion**(0.5))
def omega_minus(kappa, m1, m2, k, a):
    raiz= kappa*((m1+m2)*(m1+m2)-4.0*(m1*m2)*((np.sin(k*a/2.0))**2.0))**(0.5)
    denum=m1*m2
    num=kappa*(m1+m2)-raiz
    fraccion=num/denum
    return (fraccion**(0.5))
```

```python
#Vamos a fijar numeritos para poder graficar tranquilamente
kappa=1.0
m1=1.0
m2=5.0
a=1.0
#ahora creamos el espacio de k
k=np.linspace(-pi/a,pi/a,100)
```

```python
#Ahora calculemos los omegas
omega_mas=omega_plus(kappa, m1, m2, k, a)
omega_menos=omega_minus(kappa, m1, m2, k, a)
```

```python
#Ahora grafiquemos
figure(1, figsize=(10,10))
title("Dispercion en la zona reducida: [-pi/a;pi/a]",fontsize=18)
xlabel("k",fontsize=16)
ylabel("w",fontsize=16)
plot(k,omega_mas,label='Modo Optico, w+')
plot(k,omega_menos,label='Modo Acustico, w-')
legend(fontsize=14,bbox_to_anchor=(0.9, 0.7),bbox_transform=gcf().transFigure)
```

```python
#Ahora lo vamos a graficar para la zona extendida
#Vamos a fijar numeritos para poder graficar tranquilamente
kappa=1.0
m1=1.0
m2=5.0
a=1.0
#ahora creamos el espacio de k
k=np.linspace(-3.0*pi/a,3.0*pi/a,100)
#Ahora calculemos los omegas
omega_mas=omega_plus(kappa, m1, m2, k, a)
omega_menos=omega_minus(kappa, m1, m2, k, a)
#Ahora grafiquemos
figure(2, figsize=(10,10))
title("Dispercion en la zona extendida: [-3pi/a;3pi/a]",fontsize=18)
xlabel("k",fontsize=16)
ylabel("w",fontsize=16)
plot(k,omega_mas,label='Modo Optico, w+')
plot(k,omega_menos,label='Modo Acustico, w-')
legend(fontsize=14,bbox_to_anchor=(0.9, 0.7),bbox_transform=gcf().transFigure)
```
