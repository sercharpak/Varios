A continuación se calculará el Hessiano para cada pixel y se hayará el límite de la vesícula

```python
filas = len(Im1)
columnas = len(Im1[0])
h = 1
#Listas que guardarán las posiciones de los puntos que forman los bordes de las vesículas
xpos = []
ypos = []

#Se inicializa la matriz que generará el borde de la vesícula de acuerdo a los valores propios del Hessiano
new_img = ones((filas,columnas))

#Se recorre toda la imagen excepto en los bordes
for i in range(1,filas-1):
    for j in range(1,columnas-1):
        hessian = zeros((2,2))
        hessian[0][0] = (Im1[i+h][j]+Im1[i-j][j]-(2*Im1[i][j]))/(h**2)
        hessian[1][1] = (Im1[i][j+h]+Im1[i][j+h]-(2*Im1[i][j]))/(h**2)
        hessian[0][1] = (Im1[i+h][j+h]+Im1[i-h][j-h]+(2*Im1[i][j])-Im1[i+h][j]-Im1[i-h][j]-Im1[i][j+h]-Im1[i][j-h])/(2*h**2)
        hessian[1][0] = (Im1[i+h][j+h]+Im1[i-h][j-h]+(2*Im1[i][j])-Im1[i+h][j]-Im1[i-h][j]-Im1[i][j+h]-Im1[i][j-h])/(2*h**2)
        
        eigenvalues, eigenvectors = eig(hessian)
          
        #A continuación seleccionaremos los valores que forman el borde de la vesícula y lo guardaremos en una nueva imagen
        if(eigenvalues[0]<-1050 and eigenvalues[1]<-1050 and i>45):
            new_img[i][j] = 0
            xpos.append(i)
            ypos.append(j)
         
fig = figure(1, figsize=(10.5,8.5))
imshow(new_img[:,:], cmap = cm.Greys_r)  
#plt.title('Límite de las vesículas', fontsize=12)
```

En el paso anterior, a partir de los valores propios del Hessiano de la imagen se ubicaron los puntos que delimitan las vesículas. Esto es posible porque el Hessiano(matriz simétrica), junto con sus valores propios, da información acerca de la geometría del objeto. De acuerdo a los signos de los valores propios, se puede decidir si el punto es un mínimo o máximo relativos o un punto silla. 

Ahora realizaremos el fit que nos devolverá las coordenadas del centro del círculo y su respectivo radio.

```python
#Haciendo el fit, de acuerdo a las instrucciones de: http://nbviewer.ipython.org/github/pv/SciPy-CookBook/blob/master/ipython/Least_Squares_Circle.ipynb
x_m = mean(xpos)
y_m = mean(ypos)

#Función que calcula la definición de un círculo 
def f(beta,xpos):
    return (xpos[0]-beta[0])**2 + (xpos[1]-beta[1])**2 -beta[2]**2

def calc_r(xc,yc):
    return sqrt((xpos-xc)**2 + (ypos-yc)**2)
r_m = calc_r(x_m,y_m).mean()
beta0 = [x_m,y_m,r_m]

lsc_data  = odr.Data(row_stack([xpos, ypos]), y=1)
lsc_model = odr.Model(f, implicit=True)
lsc_odr   = odr.ODR(lsc_data, lsc_model, beta0)
lsc_out   = lsc_odr.run()

#Se obtienen las coordenadas del centro del círculo y el radio
xc_3, yc_3, R_3 = lsc_out.beta
Ri_3 = calc_r(xc_3, yc_3)
residu_3 = sum((Ri_3 - R_3)**2)
```
