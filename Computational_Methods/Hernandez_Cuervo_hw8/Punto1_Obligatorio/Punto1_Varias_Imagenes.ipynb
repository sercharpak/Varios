```python
%pylab inline
import numpy as np
import glob
from matplotlib.backends.backend_pdf import PdfPages
```

```python
#Ahora a crear circulos y jugar con esta @#%(*)@% a ver que o que
def log_likelihood(y_obs, y_model):
    chi_squared = (1.0/2.0)*sum((y_obs-y_model)**2)
    return (-chi_squared)
def my_model(r0, r1,m,phi,centro_x,centro_y,y_obs):
    theticitte=linspace(0,2*np.pi,10000)
    matricita=np.zeros((shape(y_obs)[0],shape(y_obs)[1]))
    primerPunto_x= centro_x+((np.cos(theticitte))*(r0+r1*np.sin(m*theticitte+phi)))
    primerPunto_y= centro_y+((np.sin(theticitte))*(r0+r1*np.sin(m*theticitte+phi)))
    matricita[int_(primerPunto_x),int_(primerPunto_y)]=1.0
    return matricita
```

El codigo implementado aca en Ipython funciona con cualquier imagen .tiff de vesiculas. Sin embargo se demora una cantidad significativa de tiempo para cada imagen por lo tanto en el siguiente codigo se escogeran aleatoriamente imagenes y se guardara un pdf por cada una con los histogramas y los resultados obtenidos. Los pdf con las imagenes se pueden observar en la carpeta archivosVesiculas.

```python
numero_imagenes_a_analizar=2
```

La clave aca es lograr que la primera iteracion sea la buena para que el modelo no vaya a fallar. Se pudo observar que efectivamente fallaba cuando el radio de la primera iteracion era mas grande que el radio rea, entre otras cosas.

```python
fileset = glob.glob('archivosVesiculas/*.tif')
contador=0
archivos_analizados= empty((0))
while(contador<numero_imagenes_a_analizar):
    n_aleatorio=random.randint(len(fileset))
    if(n_aleatorio not in archivos_analizados):
        archivos_analizados=append(archivos_analizados,n_aleatorio)
        contador=contador+1
        nombreArchivo=fileset[n_aleatorio]
        Imagen_Original = pylab.imread(nombreArchivo)
        #Ahora pasamos esta matriz a numeros entre 0 y 1
        maximus=amax(Imagen_Original)
        minimus=amin(Imagen_Original)
        matrizNorm=zeros((shape(Imagen_Original)[0],shape(Imagen_Original)[1]))
        matrizNorm=((1.0*Imagen_Original-minimus)/maximus)
        if(len(shape(matrizNorm))>2):
            #tiene mas de 1 capa
            matrizPromedio=zeros((shape(matrizNorm)[0],shape(matrizNorm)[1]))
            for i in range(shape(matrizNorm)[0]):
                for j in range(shape(matrizNorm)[1]):
                    matrizPromedio[i,j]=sum(matrizNorm[i][j])/len(matrizNorm[i][j])
            y_obs=matrizPromedio 
        else:
            y_obs=matrizNorm
        m_walk = empty((0)) #this is an empty list to keep all the steps
        r0_walk = empty((0))
        r1_walk = empty((0))
        phi_walk = empty((0))
        centro_X_walk = empty((0))
        centro_Y_walk = empty((0))
        l_walk = empty((0))
        nIteraciones=0
        while(nIteraciones<100):
            try:
                centro_X_walk=append(centro_X_walk, (shape(y_obs)[0]/2-(shape(y_obs)[0]/6))+random.randint(shape(y_obs)[0]*2/6))
                centro_Y_walk=append(centro_Y_walk, (shape(y_obs)[1]/2-(shape(y_obs)[1]/6))+random.randint(shape(y_obs)[1]*2/6))
                m_walk = append(m_walk, random.randint(10))
                r0_walk = append(r0_walk, sqrt(((random.randint(shape(y_obs)[0]/10))**2.0)+((random.randint(shape(y_obs)[1]/10))**2.0)))
                r1_walk = append(r1_walk, random.random())
                phi_walk = append(phi_walk, 2.0*np.pi*random.random())
                y_init = my_model(r0_walk[0], r1_walk[0], m_walk[0],phi_walk[0],centro_X_walk[0],centro_Y_walk[0],y_obs)
                nIteraciones=nIteraciones+1
                break
            except:
                nIteraciones=nIteraciones+1
                m_walk = empty((0)) #this is an empty list to keep all the steps
                r0_walk = empty((0))
                r1_walk = empty((0))
                phi_walk = empty((0))
                centro_X_walk = empty((0))
                centro_Y_walk = empty((0))
                l_walk = empty((0))
        primeraIteracion=y_init
        l_walk = append(l_walk, log_likelihood(y_obs, y_init))
        n_iterations = 60000 #this is the number of iterations I want to make
        for i in range(n_iterations):
            m_prime = np.random.normal(0, 10) 
            r0_prime = np.random.normal(r0_walk[i], 2)
            r1_prime = np.random.normal(r1_walk[i], 0.1)
            centro_X_prime = np.random.normal(centro_X_walk[i], 3)
            centro_Y_prime = np.random.normal(centro_Y_walk[i], 3)
            phi_prime = np.random.normal(phi_walk[i], 1)
            
            y_prime = my_model(r0_prime, r1_prime, m_prime,phi_prime,centro_X_prime,centro_Y_prime,y_obs)
            y_init = my_model(r0_walk[i], r1_walk[i], m_walk[i],phi_walk[i],centro_X_walk[i],centro_Y_walk[i],y_obs)
    
            l_prime = log_likelihood(y_obs, y_prime)
            l_init = log_likelihood(y_obs, y_init)
    
            alpha = exp(+(l_prime-l_init))
        
            if(alpha>=1.0):
                m_walk  = append(m_walk,m_prime)
                r1_walk  = append(r1_walk,r1_prime)
                r0_walk  = append(r0_walk,r0_prime)
                phi_walk  = append(phi_walk,phi_prime)
                centro_X_walk  = append(centro_X_walk,centro_X_prime)
                centro_Y_walk  = append(centro_Y_walk,centro_Y_prime)
                l_walk = append(l_walk, l_prime)
            else:
               beta = random.random()
               if(beta<=alpha):
                   m_walk  = append(m_walk,m_prime)
                   r1_walk  = append(r1_walk,r1_prime)
                   r0_walk  = append(r0_walk,r0_prime)
                   phi_walk  = append(phi_walk,phi_prime)
                   centro_X_walk  = append(centro_X_walk,centro_X_prime)
                   centro_Y_walk  = append(centro_Y_walk,centro_Y_prime)
                   l_walk = append(l_walk, l_prime)
               else:
                   m_walk  = append(m_walk,m_walk[i])
                   r1_walk  = append(r1_walk,r1_walk[i])
                   r0_walk  = append(r0_walk,r0_walk[i])
                   phi_walk  = append(phi_walk,phi_walk[i])
                   centro_X_walk  = append(centro_X_walk,centro_X_walk[i])
                   centro_Y_walk  = append(centro_Y_walk,centro_Y_walk[i])
                   l_walk = append(l_walk, l_init)
        max_likelihood_id = argmax(l_walk)
        best_m = m_walk[max_likelihood_id]
        best_r0 = r0_walk[max_likelihood_id]
        best_r1 = r1_walk[max_likelihood_id]
        best_phi = phi_walk[max_likelihood_id]
        best_centro_X = centro_X_walk[max_likelihood_id]
        best_centro_Y = centro_Y_walk[max_likelihood_id]
        
        histogramasPlot=plt.figure(figsize=(14,14))
        plt.subplot(421)
        count, bins, ignored =plt.hist(m_walk, 20, normed=True)
        plt.title('Histograma de m', fontsize=12)
        plt.subplot(422)
        count, bins, ignored =plt.hist(r0_walk, 20, normed=True)
        plt.title('Histograma de $r_0$', fontsize=12)
        plt.subplot(423)
        count, bins, ignored =plt.hist(r1_walk, 20, normed=True)
        plt.title('Histograma de $r_1$', fontsize=12)
        plt.subplot(424)
        count, bins, ignored =plt.hist(phi_walk, 20, normed=True)
        plt.title('Histograma de $\phi$', fontsize=12)
        plt.subplot(425)
        count, bins, ignored =plt.hist(centro_X_walk, 20, normed=True)
        plt.title('Histograma de $x_{centro}$', fontsize=12)
        plt.subplot(426)
        count, bins, ignored =plt.hist(centro_Y_walk, 20, normed=True)
        plt.title('Histograma de $y_{centro}$', fontsize=12)
        plt.subplot(427)
        plot((l_walk))
        plt.title('Error de la prueba de $\chi^{2}$')
        #Ahora grafiquemos las imagenes
        
        best_y = my_model(best_r0, best_r1, best_m,best_phi, best_centro_X, best_centro_Y, y_obs)
        
        figOriginal = figure(2, figsize=(10.5,8.5))
        imshow(y_obs[:,:], cmap = cm.Greys_r)
        plt.title(nombreArchivo+' - Imagen Normalizada', fontsize=12)
        figPrimeraIter = figure(3, figsize=(10.5,8.5))
        imshow(primeraIteracion[:,:], cmap = cm.Greys_r)
        plt.title(nombreArchivo+' -  Primera iteracion -  circulo', fontsize=12)
        figMejor = figure(4, figsize=(10.5,8.5))
        imshow(best_y[:,:], cmap = cm.Greys_r)
        plt.title(nombreArchivo+' -  Mejor Circulo con parametros \n m:' +str(best_m)+' ; $r_0$:'+str(best_r0)+' ; $r_1$:'+str(best_r1)+' ; $\phi$:'+str(best_phi)+' ; $x_{centro}$:'+str(best_centro_X)+' ; $y_{centro}$:'+str(best_centro_Y), fontsize=12)
        pp = PdfPages(nombreArchivo.strip('.tif')+'.pdf')
        pp.savefig(figOriginal)
        pp.savefig(figPrimeraIter)
        pp.savefig(figMejor)
        pp.savefig(histogramasPlot)
        pp.close()
```
