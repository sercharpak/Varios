```python
%pylab inline
import numpy as np
```

```python
def calcET(gamma,p,ro,u):
    return ((1.0/ro)*(p/(gamma-1.0))+(1.0/2.0)*(u*u))
def calcF(u,gamma):
    f=np.zeros(3)
    f[0]=u[1]
    f[1]=((u[1]*u[1])/u[0])+(gamma-1)*(u[2]-((1.0/2.0)*((u[1]*u[1])/u[0])))
    f[2]=(u[2]+(gamma-1)*(u[2]-((1.0/2.0)*((u[1]*u[1])/u[0]))))*((u[1])/u[0])
    return f
```

```python
gamma=1.4
uii=0.0
roii=1.0
pii=100.0*(10.0**3.0)
etii=calcET(gamma,pii,roii,uii)
udi=0.0
rodi=0.125
pdi=10.0*(10.0**3.0)
etdi=calcET(gamma,pdi,rodi,udi)

ti=0.0
tf=0.01
dt=0.00001
xi=-10.0
xf=10.0
n_x = 1000
#n_t = int(tf/dt
n_t = int(round(tf/dt))

x = linspace(xi, xf, n_x)
dx = x[1]-x[0]
print dt/dx
```

```python
ro_i=zeros(n_x)
vel_i=zeros(n_x)
p_i=zeros(n_x)
for i in range(n_x):
    if x[i]<0:
        ro_i[i]=roii
        p_i[i]=pii
        vel_i[i]=uii
    else:
        ro_i[i]=rodi
        p_i[i]=pdi
        vel_i[i]=udi
```

```python
uli=np.zeros(3)
uri=np.zeros(3)
fli=np.zeros(3)
fri=np.zeros(3)
```

```python
uli[0]=roii
uli[1]=roii*uii
uli[2]=roii*etii
uri[0]=rodi
uri[1]=rodi*udi
uri[2]=rodi*etdi
fli=calcF(uli,gamma)
fri=calcF(uri,gamma)
```

```python
f_i = zeros((3.0,n_x))
u_i = zeros((3.0,n_x))
for i in range (0,n_x):
    if(x[i]<0):
        u_i[0,i]=uli[0]
        u_i[1,i]=uli[1]
        u_i[2,i]=uli[2]
        f_i[0,i]=fli[0]
        f_i[1,i]=fli[1]
        f_i[2,i]=fli[2]
    else:
        u_i[0,i]=uri[0]
        u_i[1,i]=uri[1]
        u_i[2,i]=uri[2]
        f_i[0,i]=fri[0]
        f_i[1,i]=fri[1]
        f_i[2,i]=fri[2]
```

```python
u_futura=zeros((3,n_x))
u_futura[:,0]=u_i[:,0]
u_futura[:,n_x-1]=u_i[:,n_x-1]
u_medios=u_i.copy()
```

```python
for n in range(1,n_t):  # loop over time
    for i in range(1,n_x-1): #loop over space
        u_medios[:,i]=(1.0/2.0)*(u_i[:,i+1]+u_i[:,i])-(dt/(2.0*dx))*(f_i[:,i+1]-f_i[:,i])
    f_medios=zeros((3,n_x))
    for i in range(0,n_x):
        f_medios[:,i]=calcF(u_medios[:,i],gamma)
    for i in range(1,n_x-1):
        u_futura[:,i]=u_i[:,i]-(dt/dx)*(f_medios[:,i]-f_medios[:,i-1])
    f_futura=zeros((3,n_x))
    for i in range(0,n_x):
        f_futura[:,i]=calcF(u_futura[:,i],gamma)
    u_i=u_futura.copy()
    f_i=f_futura.copy()
```

```python
ro_final=u_futura[0,:]
vel_final=u_futura[1,:]/ro_final
p_final=(gamma-1)*(u_futura[2,:]-(1.0/2.0)*(u_futura[1,:]*u_futura[1,:]/u_futura[0,:]))
```
