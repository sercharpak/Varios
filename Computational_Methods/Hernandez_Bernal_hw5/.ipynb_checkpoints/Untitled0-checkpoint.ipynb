```python
%pylab inline
from scipy import fftpack
from scipy.fftpack import fft, fftfreq
```

```python
data = loadtxt("circulos.dat")
x=data[:,0]+3
y=data[:,1]+3
figure(figsize=(8,8))
scatter(x,y,s=0.01)
plt.grid()
xlabel('x')
ylabel('y')
title('Grafica original')
```

```python
delta=600
h=6.0/delta
grid=zeros((delta,delta))
for i in range (delta):
    for j in range (delta):
        count=len(x[(x<h*(i+1)) & (x>=h*(i)) & (y<h*(j+1)) & (y>=h*(j))])
        grid[i][j]=count
```

```python
figure(figsize=(8,8))
imshow(grid,cmap=cm.bone)
title('Densidad de puntos')
xlabel('x')
ylabel('y')
```

```python
freq=fftfreq(delta,h)
ft_grid=fft2(grid)
figure(figsize(8,8))
imshow(log10(abs((ft_grid))**2),cmap=cm.bone)
title('Transformada de Fourier en 2 dimensiones de funcion de densidad')
```

```python
figure(figsize(8,8))
imshow(log10(abs(fftshift(ft_grid))**2),cmap=cm.bone)
title('Transformada de Fourier reorganizada en 2 dimensiones de funcion de densidad')
```

```python
ft_x=ft_grid[:,0]
ft_y=ft_grid[:,1]
ft_r=((abs(ft_x))**2+(abs(ft_y))**2)
plot(freq,ft_r)
xlabel('Frecuencia')
ylabel('Transformada de la funcion r cuadrado')
title('Autocorrelacion de la funcion radial')
plt.grid()
```

```python
ft_r_shift=fftshift(ft_r)
freq_shift=fftshift(freq)
plot(freq_shift,ft_r_shift)
xlabel('Frecuencia')
ylabel('Transformada de la funcion r cuadrado')
title('Autocorrelacion de la funcion radial (organizada)')
plt.grid()
```

```python
ft_r_cut=empty(0)
freq_cut=empty(0)
for i in range(len(freq_shift)):
    if(abs(freq_shift[i])<=10):
        ft_r_cut=append(ft_r_cut,ft_r_shift[i])
        freq_cut=append(freq_cut,freq_shift[i])
```

```python
#Función para determinación de máximos locales
def local_maxima(xval, yval):
    xval = np.asarray(xval)
    yval = np.asarray(yval)

    sort_idx = np.argsort(xval)
    yval = yval[sort_idx]
    gradient = np.diff(yval)
    maxima = np.diff((gradient > 0).view(np.int8))
    return np.concatenate((([0],) if gradient[0] < 0 else ()) +
                          (np.where(maxima == -1)[0] + 1,) +
                          (([len(yval)-1],) if gradient[-1] > 0 else ()))
```

```python
testy= ft_r_cut[local_maxima(freq_cut,ft_r_cut)]
testx= freq_cut[local_maxima(freq_cut,ft_r_cut)]
```

```python
plot(freq_cut,ft_r_cut)
scatter(testx,testy)
xlabel('Frecuencia')
ylabel('Transformada de la funcion r cuadrado')
title('Autocorrelacion de la funcion radial (organizada)')
legend(['Transformada','Maximos locales'])
plt.grid()
```

```python
half=len(ft_r_cut)/2
new=empty(0)
new2=empty(0)
for i in range(half):
    new=np.append(new,(ft_r_cut[half+i]+ft_r_cut[half-i])/2)
    new2=np.append(new2,(freq_cut[half+i]))
plot(new2,new)
testy= new[local_maxima(new2,new)]
testx= new2[local_maxima(new2,new)]
scatter(testx,testy)
xlim(0.01,10)
ylim(0,1000000)
plt.grid()
xlabel('Frecuencia')
ylabel('Transformada promediada')
title('Autocorrelacion de la funcion radial en la parte positiva')
legend(['Transformada','Maximos locales'])
```

```python
dx=empty(0)
deltax=empty(0)
count=empty(0)
for i in range(len(testx)-1):
    dx=np.append(dx,testx[i+1]-testx[i])
for i in range(len(dx)):
    if dx[i]!=deltax.all():
        deltax=np.append(deltax,dx[i])
        count=np.append(count,0)
    count[deltax==dx[i]]+=1
scatter(deltax,count)
ylim(0,None)
xlabel('$\Delta$ de frecuencias')
ylabel('Frecuencia de aparicion')
title('Histograma de frecuencia de aparicion de los $\Delta$')
plt.grid()
```

```python
radios=1/(2*deltax)
scatter(radios,count)
xlabel('Radio de los circulos')
ylabel('Frecuencia de aparicion')
title('Histograma de frecuencia de aparicion de los radios')
plt.grid()
```
