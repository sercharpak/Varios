### Punto 2 - White Noise vs Pink Noise

En este punto se grabó la voz de una compañera, y se generó ruido blanco y ruido rosa con esta señal de entrada. La generación del ruido blanco se hizo de acuerdo a lo visto en el curso. La generación de ruido rosa se hizo tomando como guía el código y la discusión en esta página: http://www.firstpr.com.au/dsp/pink-noise/

```python
%pylab inline
import scikits.audiolab as audio
from scipy.fftpack import fft, fftfreq
from __future__ import division
import matplotlib.pyplot as plt
import numpy as np
from numpy import pi
```

```python
input_signal, sampling_rate, enc = audio.wavread("voiceBarbara.wav")
```

```python
time_array = arange(0, len(input_signal)/float(sampling_rate), 1/float(sampling_rate))
```

```python
print len(time_array), len(input_signal)
```

Luego de escuchar a Bárbara se puede pasar generar el ruido blanco (y luego el ruido rosa). Empezamos con realisar la transformación de Fourier a la señal original.

```python
n_points=len(input_signal)
fft_input_signal = fftshift(fft(input_signal) / n_points )
freq = fftshift(fftfreq(n_points, 1) )
plot(freq,abs(fft_input_signal))
xlabel('Frecuencia [hz]')
ylabel('Transformada de la senal')
title('Transformada de Fourier de senal original')
plt.grid()
```

Ahora generamos el ruido blanco en el espacio de Fourier. Dividimos la transformada por su valor absoluto y luego multiplicamos el todo por un número arbritrario (en este caso 5).

```python
fft_white=5*(fft_input_signal/np.abs(fft_input_signal))
```

```python
scatter(freq,log10(np.absolute(fft_input_signal)**2))
xlabel('Frecuencia [hz]')
ylabel('Espectro de potencias')
title('Espectro de potencias senal original')
plt.grid()
```

Podemos ahora observar el espectro de potencias esperado del ruido blanco.

```python
scatter(freq, log10(np.absolute(fft_white)**2)) 
xlabel('Frecuencia [hz]')
ylabel('Espectro de potencias')
title('Espectro de potencias ruido blanco')
plt.grid()
```

Ahora realizamos la transformación inversa y nos preparamos para oir al ruido blanco de la voz de Bárbara.

```python
white=ifft(fft_white)
plt.plot(real(white), linewidth=4, color='red')
plt.plot(input_signal)
xlabel('Tiempo [s]')
ylabel('Amplitud')
title('Senal original y ruido blanco')
legend(['Ruido blanco','Senal original'])
plt.grid()
```

```python
output_file = audio.Sndfile("Voz_Blanco.wav", 'w', audio.Format('wav'), 1, sampling_rate)
output_file.write_frames(real(white))
output_file.close()
```

Una vez que oímos el ruido blanco de Bárbara, podemos prepararnos a generar el ruido rosa. En el espacio de Fourier, la transformada de Fourier del ruido rosa se obtiene dividiendo la transformada del ruido blanco (que es obtuvo anteriormente) por las frecuencias elevadas a un parámetro $\alpha$ (con $0 < \alpha < 2$). Efectivamente el ruido rosa se define como una señal con un espectro de potencias inversamente proporcional a sus frecuencias elevadas a un $\alpha$ (con $0 < \alpha < 2$). Acá se divide por 1000 para simplificar el tratamiento de la señal.

```python
alpha=1
fft_pink=(fft_white)/(abs(freq)**alpha)/(10**3)
half_n=n_points/2
fft_pink[half_n]=0
plot(freq,fft_pink)
xlabel('Frecuencia [hz]')
ylabel('Transformada del ruido rosa')
title('Transformada de Fourier del ruido rosa')
plt.grid()
```

Podemos ahora observar el espectro de potencias del ruido rosa.

```python
scatter((freq),log10(np.absolute(fft_pink)**2))
xlabel('Frecuencia [hz]')
ylabel('Espectro de potencias')
title('Espectro de potencias ruido rosa')
plt.grid()
```

Al hacer la transormación inversa de Fourier y observar las gráficas de la señal rosa y la señal original, podemos observar que la amplitud de la señal rosa es substancialmente más pequeña que la señal original (será por lo tanto más delicada de escuchar).

```python
pink=ifft(fft_pink)
plt.plot(real(pink), linewidth=4, color='red')
plt.plot(input_signal)
xlabel('Tiempo [s]')
ylabel('Amplitud')
title('Senal original y ruido rosa')
legend(['Ruido rosa','Senal original'])
plt.grid()
```

```python
output_file = audio.Sndfile("Voz_Rosa.wav", 'w', audio.Format('wav'), 1, sampling_rate)
output_file.write_frames(real(pink)*100)
output_file.close()
```
