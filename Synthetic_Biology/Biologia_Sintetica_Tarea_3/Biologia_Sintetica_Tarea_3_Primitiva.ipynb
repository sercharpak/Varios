```python
#Tasa de transcripcion constante
#Tasa de traduccion constante por mRNA
#Dilucion por Proteina y Destruccion por mRNA constantes (r_punto=k_r -gamma_r y p_punto=k_p*r-gamma_p*p
#Aproximacion de intervales de tiempo delta_t pequenos con tasas constantes de creacion y destruccion por intervalo.
#Escoger constantes biologicamente relevantes
```

```python
#y mirar que tipo de evento son.
def evento_rna(num_rna,delta_t):
    e1=np.random.random()
    e2=np.random.random()
    if((e1<(k_r*delta_t)) and (e2>=(gamma_r*num_rna*delta_t))):
        num_rna=num_rna+1.0
    elif((e1>=(k_r*delta_t)) and (e2<(gamma_r*num_rna*delta_t))):
        if(num_rna>0.0):
            num_rna=num_rna-1.0
    return num_rna
def evento_prot(num_rna,num_prot,delta_t):
    e1=np.random.random()
    e2=np.random.random()
    if((e1<(k_p*num_rna*delta_t)) and (e2>=(gamma_p*num_prot*delta_t))):
        num_prot=num_prot+1.0
    elif((e1>=(k_p*num_rna*delta_t)) and (e2<(gamma_p*num_prot*delta_t))):
        if(num_prot>0.0):
            num_prot=num_prot-1.0
    return num_prot
```

```python
#Constantes biologicamente relevantes
k_r=1.0 #rna/min
k_p=50.0 # 1 prot/(min*rna)
K=0.5 #cte de retroalimentacion
gamma_p=1.0/30.0 #  prot/min
gamma_r=1.0/5.0 #  rna/min
```

```python
#Probemos entonces
delta_t=0.005
t_limite=1000
num_points=int(t_limite/delta_t)
t_walk=np.zeros(num_points)
r_walk=np.zeros(num_points)
prot_walk=np.zeros(num_points)
for i in range (1,num_points):
    t_walk[i]=t_walk[i-1]+delta_t
    r_walk[i]=evento_rna(r_walk[i-1],delta_t)
    prot_walk[i]=evento_prot(r_walk[i-1],prot_walk[i-1],delta_t)
```

```python
#Graficar
plt.figure(1)
plt.plot(t_walk,r_walk)
plt.title("Numero de Rnas vs tiempo")
xlabel("tiempo (min)")
ylabel("Numero de Rnas ")

plt.figure(2)
plt.plot(t_walk,prot_walk)
plt.title("Numero de proteinas vs tiempo")
xlabel("tiempo (min)")
ylabel("Numero de proteinas ")
```
