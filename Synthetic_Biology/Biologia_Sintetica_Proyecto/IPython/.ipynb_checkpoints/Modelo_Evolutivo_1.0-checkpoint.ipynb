Vamos a realizar un analisis de poblaciones de las bacterias.

##Constantes

Encontrar constantes (valores numericos) es todo un reto. En efecto para encontrar estas constantes se debe realizar un importante trabajo experimental y, muchas veces, este no ha sido realizado aun.

Nuestras constante seran entonces, en un primero momento, proporcionales a las constantes de las bacterias (fitness) encontradas en la literatura.

### E. Coli
Para E.Coli se encontro en la literatura que modificada con un plasmido R1 tiene un fitness de 2.051±0.087, mientras que sin dicho plasmido tiene un fitness de 1.0 [1].

```python
#E Coli
fit_Ecol=2.051
```

### Salmonella Enterica
Para Salmonella Enterica. se encontro en la literatura que modifcada con un plasmido R1 tiene un fitness de 1.558±0.059 mientras que sin dicho plasmido tiene un fitness de 1.000±0.074.[1]

```python
#Salmonella
fit_Salm=1.558
```

### R. Rhodnii
Para R.Rhodnii se encontro en la literatura que tiene un fitness [1]

```python
#Rhodnii
fit_Rhod=1.2
```

### Control
Vamos a tomar el control con fitness normalizado de 1.

```python
#Control
fit_Cont=1.0
```

##Ahora se define las condiciones iniciales
Definimos que todas las poblaciones iniciales seran de 45 vectores

```python
#Poblaciones Iniciales
num_vec_Ecol=45
num_vec_Salm=45
num_vec_Rhod=45
num_vec_Cont=45
```

```python
def evento_gillepsie(num_vec_Ecol,num_vec_Salm,num_vec_Rhod,num_vec_Cont,t,fit_Ecol,fit_Salm,fit_Rhod,fit_Cont):
    e1=np.random.random() #se crea un numero aleatorio entre 0 y 1 mediante una distribucion uniforme 
    #Normalicemos las constantes
    k_total=fit_Ecol+fit_Salm+fit_Rhod+fit_Cont
    paso_t=np.random.exponential()/(k_total)
    t_nuevo=t+paso_t
    #Ahora mimremos que tipo de evento sucedio
    if(e1<(fit_Ecol/k_total)):#Habemus Vec_E_Coli
        num_vec_Ecol_new=num_vec_Ecol+1.0
        num_vec_Salm_new=num_vec_Salm
        num_vec_Rhod_new=num_vec_Rhod
        num_vec_Cont_new=num_vec_Cont
    elif((e1<((fit_Ecol+fit_Salm)/k_total)) and (e1>=(fit_Ecol/k_total))):#Habemus Vec_Salm
        num_vec_Salm_new=num_vec_Salm+1.0
        num_vec_Ecol_new=num_vec_Ecol
        num_vec_Rhod_new=num_vec_Rhod
        num_vec_Cont_new=num_vec_Cont
    elif((e1<((fit_Ecol+fit_Salm+fit_Rhod)/k_total)) and (e1>=((fit_Ecol+fit_Salm)/k_total))):#Habemus Vec_Rhodn
        num_vec_Rhod_new=num_vec_Rhod+1.0
        num_vec_Ecol_new=num_vec_Ecol
        num_vec_Salm_new=num_vec_Salm
        num_vec_Cont_new=num_vec_Cont
    else:#Habemus Vec_Cont
        num_vec_Cont_new=num_vec_Cont+1.0
        num_vec_Ecol_new=num_vec_Ecol
        num_vec_Rhod_new=num_vec_Rhod
        num_vec_Salm_new=num_vec_Salm
    return num_vec_Ecol_new, num_vec_Salm_new,num_vec_Rhod_new,num_vec_Cont_new,t_nuevo

def simulacion_gillepsie(num_corridas,t_limite,delta_t,fit_Ecol,fit_Salm,fit_Rhod,fit_Cont):
    contador_corridas=0.0
    num_puntos=int(t_limite/delta_t)
    matriz_vec_Ecol=np.zeros((num_corridas,num_puntos))
    matriz_vec_Salm=np.zeros((num_corridas,num_puntos))
    matriz_vec_Rhod=np.zeros((num_corridas,num_puntos))
    matriz_vec_Cont=np.zeros((num_corridas,num_puntos))
    
    while (contador_corridas<num_corridas):
        num_vec_Ecol=45
        num_vec_Salm=45
        num_vec_Rhod=45
        num_vec_Cont=45
        t_total=0.0
        t_walk=[]
        vec_Ecol_walk=[]
        vec_Salm_walk=[]
        vec_Rhod_walk=[]
        vec_Cont_walk=[]
        while (t_total<t_limite):
            num_vec_Ecol, num_vec_Salm,num_vec_Rhod,num_vec_Cont,t_total=evento_gillepsie(num_vec_Ecol,num_vec_Salm,num_vec_Rhod,num_vec_Cont,t_total,fit_Ecol,fit_Salm,fit_Rhod,fit_Cont)
            t_walk.append(t_total)
            vec_Ecol_walk.append(num_vec_Ecol)
            vec_Salm_walk.append(num_vec_Salm)
            vec_Rhod_walk.append(num_vec_Rhod)
            vec_Cont_walk.append(num_vec_Cont)
        #Ahora se deben reformar los arreglos por los tiempos
        t_final = np.linspace(0, t_limite-delta_t, t_limite/delta_t)
        vec_Ecol_final = np.zeros(len(t_final))
        vec_Salm_final = np.zeros(len(t_final))
        vec_Rhod_final = np.zeros(len(t_final))
        vec_Cont_final = np.zeros(len(t_final))
        pos=0
        for i in range (len(t_final)):
            while((t_walk[pos] < t_final[i])&((pos+1)<len(vec_Ecol_walk))):
                pos=pos+1
            vec_Ecol_final[i] = vec_Ecol_walk[pos]
            vec_Salm_final[i] = vec_Salm_walk[pos]
            vec_Rhod_final[i] = vec_Rhod_walk[pos]
            vec_Cont_final[i] = vec_Cont_walk[pos]
        matriz_vec_Ecol[contador_corridas]=vec_Ecol_final
        matriz_vec_Salm[contador_corridas]=vec_Salm_final
        matriz_vec_Rhod[contador_corridas]=vec_Rhod_final
        matriz_vec_Cont[contador_corridas]=vec_Cont_final
        contador_corridas=contador_corridas+1.0
    
    #Ahora calculamos la media y la desviacion
    vec_Ecol_mean=np.zeros(num_puntos)
    vec_Ecol_std=np.zeros(num_puntos)
    vec_Salm_mean=np.zeros(num_puntos)
    vec_Salm_std=np.zeros(num_puntos)
    vec_Rhod_mean=np.zeros(num_puntos)
    vec_Rhod_std=np.zeros(num_puntos)
    vec_Cont_mean=np.zeros(num_puntos)
    vec_Cont_std=np.zeros(num_puntos)
    for i in range(num_puntos):
        vec_Ecol_mean[i]=np.mean(matriz_vec_Ecol[:,i])
        vec_Ecol_std[i]=np.std(matriz_vec_Ecol[:,i])
        vec_Salm_mean[i]=np.mean(matriz_vec_Salm[:,i])
        vec_Salm_std[i]=np.std(matriz_vec_Salm[:,i])
        vec_Rhod_mean[i]=np.mean(matriz_vec_Rhod[:,i])
        vec_Rhod_std[i]=np.std(matriz_vec_Rhod[:,i])
        vec_Cont_mean[i]=np.mean(matriz_vec_Cont[:,i])
        vec_Cont_std[i]=np.std(matriz_vec_Cont[:,i])
        
    return t_final, matriz_vec_Ecol, matriz_vec_Salm, matriz_vec_Rhod, matriz_vec_Cont, vec_Ecol_mean, vec_Ecol_std, vec_Salm_mean,vec_Salm_std,vec_Rhod_mean,vec_Rhod_std,vec_Cont_mean,vec_Cont_std
```

```python
#num_corridas
num_corridas=100
delta_t=0.001 #min
t_limite=100.0 #min
```

```python
t_ini=time.time()
t_final, matriz_vec_Ecol, matriz_vec_Salm, matriz_vec_Rhod, matriz_vec_Cont, vec_Ecol_mean, vec_Ecol_std, vec_Salm_mean,vec_Salm_std,vec_Rhod_mean,vec_Rhod_std,vec_Cont_mean,vec_Cont_std = simulacion_gillepsie(num_corridas,t_limite,delta_t,fit_Ecol,fit_Salm,fit_Rhod,fit_Cont)
t_fin=time.time()
print "Tiempo transcurrido: "+str((t_fin-t_ini))
```
